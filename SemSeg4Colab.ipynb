```
from scipy.spatial.distance import directed_hausdorff
import numpy as np
import os
# !pip install SimpleITK
import SimpleITK as sitk

path2data = '/content/gdrive/My Drive/TFG/TFG MARINA CALZADA/Seg_github/Semantic-Segmentation-Suite/Test_Binary/'

def load_itk_image(filename):
    itkimage = sitk.ReadImage(filename)
    numpyImage = sitk.GetArrayFromImage(itkimage)
    return numpyImage

HM=np.zeros(172)
i=0
for file in os.listdir(path2data):
  if file.endswith('_gt.png'):
    filename=file.replace('gt.png','')
    frame = load_itk_image(path2data + filename + 'pred.png')[:,:,1]
    gt = load_itk_image(path2data + filename + 'gt.png') [:,:,1]
    HM [i] = directed_hausdorff(frame, gt)[0]
    # if HM[i]==0:
    #   sub=gt-frame
    #   tot=sub.sum()
    #   print(tot)
    i +=1
    
mean_HM=np.mean(HM)
print(mean_HM)
```
