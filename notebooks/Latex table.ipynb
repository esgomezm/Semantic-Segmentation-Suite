```
path2data='/content/drive/My Drive/TFG/TFG MARINA CALZADA/Seg_github/eval_data_server/'
path2cell = '/content/drive/My Drive/TFG/TFG MARINA CALZADA/Seg_github/eval_data_server/Results_CellSegmentation/'
path2cell = '/content/drive/My Drive/TFG/TFG MARINA CALZADA/Seg_github/eval_data_server/Results_CellSegmentation/'
vid_order = pd.read_csv(path2cell + 'videos_name.csv', sep=';', header=0)
# vid_cell = pd.read_csv(path2cell + 'results.csv', sep=';', header=0)
# vid_det = pd.read_csv(path2cell + 'results_det.csv', sep=';', header=0)
vid_cell = pd.read_csv(path2cell + 'results_dense.csv', sep=';', header=0)
vid_det = pd.read_csv(path2cell + 'results_det_dense.csv', sep=';', header=0)

df1 = pd.DataFrame(vid_order[['Video', 'Name']])

df1['DET_Reduce']='0'
for i in range(len(df1['Video'])):
    for j in range(len(vid_det['Video'])):
        if df1['Video'][i]==vid_det['Video'][j]:
            df1['DET_Reduce'][i]=vid_det['dense_Reduce'][j]
            

df1['Seg_reduce']='0'
for i in range(len(df1['Video'])):
    for j in range(len(vid_cell['Video'])):
        if df1['Video'][i]==vid_cell['Video'][j]:
            df1['Seg_reduce'][i]=vid_cell['dense_Reduce'][j]


df1['DET_Constante']='0'
for i in range(len(df1['Video'])):
    for j in range(len(vid_det['Video'])):
        if df1['Video'][i]==vid_det['Video'][j]:
            df1['DET_Constante'][i]=vid_det['dense_cons'][j]    

df1['Seg_Constante']='0'
for i in range(len(df1['Video'])):
    for j in range(len(vid_cell['Video'])):
        if df1['Video'][i]==vid_cell['Video'][j]:
            df1['Seg_Constante'][i]=vid_cell['dense_cons'][j]
            

df1['Model5CSB']='0'
for i in range(len(df1['Video'])):
        value=(np.float64(df1['DET_Constante'][i])+df1['Seg_Constante'][i])/2
        df1['Model5CSB'][i]='$'+str(value)[:4]+'$'
        # df1['Model5CSB'][i]=value

df1['Model6CSB']='0'
for i in range(len(df1['Video'])):
    val = (np.float64(df1['DET_Reduce'][i])+np.float64(df1['Seg_reduce'][i]))/2
    df1['Model6CSB'][i]='$'+str(val)[:4]+'$'
    # df1['Model6CSB'][i]=val


# df = pd.DataFrame(df1)

# df.mean(axis = 0) 


print(df1.to_latex(index=False, columns=['Name', 'DET_Constante','Seg_Constante','Model5CSB','DET_Reduce', 'Seg_reduce','Model6CSB' ]))
```
